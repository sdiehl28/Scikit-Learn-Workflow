# Lahman and Retrosheet Data Dictionary

A data dictionary is sometimes called a "codebook".  Essentially it is a mapping between field name and field description.

Most of the field names used in Lahman and Retrosheet are close to the standard MLB baseball abbreviation.  Two good baseball glossaries, which provide abbreviations and descriptions, are:
* http://www.espn.com/gen/editors/mlb/glossary.html  
* https://baseballtips.com/glossary/index.html/ 

## Lahman Data Dictionary

This is well described by:  
http://www.seanlahman.com/files/database/readme2016.txt  

The main difference is that the csv file MASTER has been renamed to People.

The csv files for Lahman are "tidy", to use the Data Science term, or "normalized" to use a database term.

Tidy data is much easier to work with.  The data created by the Retrosheet parsers is not tidy, but will be made tidy along the lines of the Lahman data, prior to analysis.

## Retrosheet Data Dictionary

The csv files for Retrosheet are generated by parsing the Retrosheet play by play files (aka event files).

Each parser generates a header row with a field name for each column.  This header row is used to create the default field names by Pandas read_csv().

The parsers are the open source parsers created by Dr. T. L. Turocy.  
Parser Description: http://chadwick.sourceforge.net/doc/index.html  
Parser Source: https://sourceforge.net/projects/chadwick/  

Note that as of March 2019, the cwdaily parser, published in July 2018, is not described on the above links, however it is very similar to Dr. Turocy's other parsers. 

These open source parsers replace the closed-source Windows only parsers named BEVENT.exe and BGAME.exe, which are described on the Retrosheet website but are not used in this study.

The data dictionary for Retrosheet can be found by running the parser's '-d' option to get a description of each field, and by running the '-n' option to get the name of each field.  These can be combined to produce the data dictionary.

Alternatively, and not necessarily recommended in general but used in a later notebook, the C source code for the parser can itself be parsed using regular expressions to derive the data dictionary.

## Tidy Baseball Data

In the following the terms, "table", "csv file", "DataFrame" will be used interchangeably.

Lahman has three primary tables aggregated per year per player:
* batting
* fielding
* pitching

Additionally there is a table aggregated per year per team:
* teams

The retrosheet data will be made tidy in a similar fashion.

cwdaily produces a single csv file that will be broken up as:
* b_player_game: batting attributes per player per game
* p_player_game: pitching attributes per player per game
* f_player_game: fielding attributes per player per game

cwgame produces a single csv file that will be broken up as:
* game: game attributes (time, attendance, etc.)
* b_team_game: batting attributes per team per game
* p_team_game: pitching attributes per team per game
* f_team_game: fielding attributes per team per game
